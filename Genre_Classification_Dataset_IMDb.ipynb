```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,accuracy_score
```

```python
train_file="train_data.txt"
```

```python
df=pd.read_csv(train_file, sep=" ::: ", header=None, names=["ID", "TITLE", "GENRE", "DESCRIPTION"], engine='python')
```

```python
df.head(3)
```

```python
df=df.drop(columns=["ID"])
```

```python
df.head(2)
```

```python
x_train=df["DESCRIPTION"]
y_train=df["GENRE"]
```

```python
tfidf = TfidfVectorizer(stop_words="english")
```

```python
x_train_tfidf = tfidf.fit_transform(x_train)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(x_train_tfidf, y_train)
```

```python
test_file="test_data.txt"
```

```python
test_df=pd.read_csv(test_file, sep=" ::: ", header=None, names=["ID", "TITLE", "DESCRIPTION"], engine='python')
```

```python
test_df=test_df.drop(columns=["ID"])
```

```python
x_test_tfidf=tfidf.transform(test_df["DESCRIPTION"])
```

```python
y_pred = model.predict(x_test_tfidf)
```

```python
test_solution_file = "test_data_solution.txt"
```

```python
test_solution_df = pd.read_csv(
    test_solution_file,
    sep=" ::: ",
    header=None,
    names=["ID","TITLE", "GENRE","DESCRIPTION"],
    engine="python"
)
y_test = test_solution_df["GENRE"].str.strip().str.lower()
```

```python
test_solution_df.head(5)
```

```python
y_pred = pd.Series(y_pred).str.strip().str.lower()

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
